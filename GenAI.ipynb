**What is a DataStore?**

A DataStore is a cloud data storage or database which is leveraged by GenAI models to retrieve, generate, and analyze data.

They can be your text files, pdfs, documents etc.



**What is RAG pipeline?**

Imagine you ask a smart AI assistant a question, but instead of just guessing, it looks up real information first before answering.


Here’s how it works in 5 simple steps:

1️⃣ You ask a question → Like "How do I fix my WiFi?"

2️⃣ AI searches for answers → It checks help articles, past solutions, and FAQs.

3️⃣ AI reads the best info → It picks the most relevant details.

4️⃣ AI writes an answer → It combines the found info with its own knowledge.

5️⃣ You get a solid response → More accurate than AI making things up! ✅

**What are Playbook/Agents?**

These are AI workflows with set of instructions for AI to think, plan, reason and take action.

Example: AI Travel Assistant (Agentic Agent) ✈️

Let’s say you want an AI to plan a vacation.


🔹 A basic AI powered search engine will provide you list of your flights with filters.

🔹 An agentic agent would:

1️⃣ Search flights & hotels based on your budget.

2️⃣ Compare prices & suggest the best deals.

3️⃣ Book everything for you (if given permission).

4️⃣ Create an itinerary with weather updates & local
attractions.


5️⃣ Adjust plans if flights change or prices drop.
